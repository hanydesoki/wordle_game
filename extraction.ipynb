```python
import pandas as pd
import numpy as np
import string
import unicodedata
#!pip install pyxlsb
```

```python
SOURCE_FILE_NAME = "Lexique383.xlsb"
ALL_LETTERS = list(string.ascii_lowercase + string.ascii_uppercase)
```

```python
df = pd.read_excel(SOURCE_FILE_NAME)
```

```python
df.head(20)
```

```python
def strip_accents(text):
    try:
        text = unicode(text, 'utf-8')
    except NameError: # unicode is a default on python 3 
        pass
    
    try:
        text = unicodedata.normalize('NFD', text)\
               .encode('ascii', 'ignore')\
               .decode("utf-8")
    except TypeError:
        return ""

    return str(text)
```

```python
def is_valid_word(word):
    try:
        for c in word:
            if c not in ALL_LETTERS:
                return False

        if len(word) > 11 or len(word) < 4:
            return False
    except:
        return False
    
    return True
```

```python
df['words_without_accents'] = df['1_ortho'].map(strip_accents)
```

```python
df.head(20)
```

```python
df_cleaned = df[df['words_without_accents'].map(is_valid_word)]
df_cleaned['len_word'] = df['words_without_accents'].map(len)
```

```python
df['4_cgram'].value_counts()
```

```python
df_cleaned['words_without_accents'].dropna(inplace=True)
```

```python
df_cleaned['infinitif'] = ((df_cleaned['3_lemme'] == df_cleaned['words_without_accents']) & (df_cleaned['4_cgram'] == 'VER'))
```

```python
df_cleaned['freqs'] = (df_cleaned['7_freqlemfilms2'] + df_cleaned['8_freqlemlivres'] + df_cleaned['9_freqfilms2'] + df_cleaned['10_freqlivres'] )/ 4
```

```python
df_cleaned.head(20)
```

```python
df_cleaned_common = df_cleaned[(df_cleaned['freqs'] > 1) & ((df_cleaned['4_cgram'] != 'VER') | ((df_cleaned['4_cgram'] == 'VER') & (df_cleaned['infinitif'] == True)))]
df_cleaned_common = df_cleaned_common[df_cleaned_common['5_genre'].isin([np.nan, 'm']) & df_cleaned_common['6_nombre'].isin([np.nan, 's']) & (df_cleaned['4_cgram'] != 'AUX')]
```

```python
df_cleaned_common.head()
```

```python
for len_word in df_cleaned['len_word'].unique():
    df_filtered = df_cleaned[df_cleaned['len_word'] == len_word]
    df_filtered_common = df_cleaned_common[df_cleaned['len_word'] == len_word]
    
    words = pd.Series(df_filtered['words_without_accents'].unique())
    common_words = pd.Series(df_filtered_common['words_without_accents'].unique())
    
    with open(f'Words/francais_{len_word}.txt', 'w') as f:
        dfAsString = words.to_string(header=False, index=False)
        f.write(dfAsString)
        
    with open(f'Words/francais_common_{len_word}.txt', 'w') as f:
        dfAsString = common_words.to_string(header=False, index=False)
        f.write(dfAsString)
```
